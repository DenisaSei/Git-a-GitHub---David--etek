```python
# definice funkce
def sum_numbers(cislo_1, cislo_2, cislo_3):
  return cislo_1 + cislo_2 + cislo_3
```

```python
#volání funkce
sum_numbers(5, 10, 20)
```

```python
def preloz_vino(typ_vina):
  if typ_vina == "white":
    return "bílé"
  elif typ_vina == "red":
    return "červené"
  else:
    print("neznám")
    return ""
```
